```python
import os
import json
# load roadnet
path_to_roadnet_input = os.path.join(
    "manual_16x3", "roadnet_16_3.json"
)
path_to_route_file = os.path.join("manual_16x3", "optional_routes.json")
path_to_roadnet_output = os.path.join(
    "manual_16x3", "leftflow.json"
)
roadnet_json = json.load(open(path_to_roadnet_input, "r"))
route = json.load(open(path_to_route_file, "r"))
roads = roadnet_json["roads"]
intersections = roadnet_json["intersections"]
```

```python
start_road = []
for road in roads:
    id = road['id'].split("_")
    if (id[1]==id[3]=='0'):
        start_road.append(road['id'])
```
